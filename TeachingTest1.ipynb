```python
import numpy as np

def ishigami(x1, x2, x3, a=7, b=0.1):
    return np.sin(x1) + a * np.power(np.sin(x2), 2) + b * np.power(np.sin(x3),4)

x1 = (np.random.rand(int(1e5))-0.5)*2*np.pi
x2 = (np.random.rand(int(1e5))-0.5)*2*np.pi
x3 = np.zeros(x1.shape)

y = ishigami(x1,x2,x3)

print(y)
```

```python
import seaborn as sns

sns.set_style("whitegrid")

sns.scatterplot(x= x1, y=y, s=0.1)

```

```python
sns.scatterplot(x=x2, y=y, s=0.1)
```
